```python
def fun_view():
    
    null_values = df1.apply(lambda x:x.isnull().sum())
    
    blank_char = df1.apply(lambda x:x.isin(['?']).sum())
    
    percent_blank_char = df1.apply(lambda x:round((x.isin(['?']).sum()/df1.shape[0])*100, 2))
    
    unique_values = df1.apply(lambda x:len(x.unique()))
    
    return pd.DataFrame({'null_values':null_values,
                         '? Values':blank_char,'% ? Values':percent_blank_char
                        ,'unique_values':unique_values})

print('Function Information',fun_view.__doc__)
view = fun_view()
display(view)
```

```python
df['class'] = df['class'].astype(int)
df['class'].value_counts()
```

```python
df.head()
```

```python
x = df.drop('class', axis = 1)
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
from imblearn.over_sampling import SMOTE
s=SMOTE(ratio='minority')
x_train,y_train=s.fit_sample(x_train,y_train)
x_test,y_test=s.fit_sample(x_test,y_test)
print(sum(y_train==1))
print(sum(y_train==0))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb_pred = xgb.fit(x_train, y_train).predict(x_test)
print('Training score is',xgb.score(x_train, y_train))
print('Test score is ',xgb.score(x_test, y_test))
print('Accuracy is',accuracy_score(y_test, xgb_pred))
```

```python
print(classification_report(y_test, xgb_pred))
```

```python
print(confusion_matrix(y_test, xgb_pred))
```
